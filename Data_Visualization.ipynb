```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# **Scatter Plot**

**Use When**: You want to visualize the relationship between two continuous variables.

**How to Read**:

Points represent individual data observations.

The position of each point on the x and y axes represents the values of the two variables.

Color (hue) can be used to represent different categories.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Scatter Plot
plt.figure(figsize=(5, 3))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

# **Line Plot**

**Use When**: You want to visualize trends over time or ordered categories.


**How to Read**:

Points represent data observations.

Lines connect data points in order.

Useful for identifying trends and changes over time.

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
# Line Plot
plt.figure(figsize=(6, 4))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot of Signal over Time")
plt.show()
```

# **Bar Plot**

**Use When**: You want to compare quantities across different categories.

**How to Read**:

The height of the bars represents the value of the variable.

Categories are represented on the x-axis.

Useful for comparing the magnitude of different categories.

```python
# tips data
plt.figure(figsize=(6, 4))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Bar Plot of Total Bill by Day and Sex")
plt.show()
```

# **Histogram**

**Use When:** You want to understand the distribution of a single continuous variable.

**How to Read:**

The x-axis represents the value range of the variable.

The y-axis represents the frequency of observations in each bin.

The shape of the histogram shows the distribution of the data.

```python
# tips data
plt.figure(figsize=(6, 4))
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

# **Box Plot**

**Use When:** You want to visualize the distribution of a variable and identify outliers.

**How to Read:**

The box represents the interquartile range (IQR).

The line inside the box represents the median.

Whiskers extend to the minimum and maximum values within 1.5 * IQR.

Points outside the whiskers are potential outliers.

```python
# tips data
plt.figure(figsize=(6, 4))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot of Total Bill by Day and Sex")
plt.show()
```

# **Heatmap**

**Use When:** You want to visualize data in a matrix form, showing the magnitude of values with color.

**How to Read:**

Each cell in the heatmap represents a data point.

The color of the cell represents the value of the data point.

Useful for identifying patterns and correlations in large datasets.

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
flights_pivot.head()
```

```python
# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap of Monthly Flights")
plt.show()
```

# **Combining Plots**

Sometimes, combining multiple plots can provide deeper insights. For example, using pair plots to visualize relationships between multiple pairs of variables.

**How to Read:**

Diagonal shows the distribution of each variable.

Off-diagonal shows scatter plots of variable pairs.

Useful for visualizing relationships between multiple variables.

```python
#tips data
# Pair Plot
sns.pairplot(tips, hue="sex")
plt.show()
```

# **Customizing Plots**

Seaborn and Matplotlib offer extensive customization options to make your plots more informative and visually appealing.

Customization Options:

Titles and labels: plt.title(), plt.xlabel(), plt.ylabel()

Legends: plt.legend()

Grid: plt.grid(True)

```python
# Customizing Plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Customized Scatter Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Time of Day")
plt.grid(True)
plt.show()
```
